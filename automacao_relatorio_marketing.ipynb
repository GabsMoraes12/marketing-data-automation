```python
import pandas as pd
import openpyxl
from openpyxl.chart import BarChart, Reference
from openpyxl.styles import Font, PatternFill

dados_brutos = {
    'Campanha': ['Black Friday', 'Institucional', 'Lançamento Curso', 'Remarketing', 'Lead Generation'],
    'Investimento (R$)': [5000, 1200, 8000, 2500, 3000],
    'Retorno (R$)': [15000, 800, 24000, 10000, 4500]
}

df = pd.DataFrame(dados_brutos)

df['ROI'] = (df['Retorno (R$)'] / df['Investimento (R$)']).round(2)

nome_arquivo = 'Relatorio_Performance_Marketing.xlsx'
df.to_excel(nome_arquivo, index=False, sheet_name='Métricas')

wb = openpyxl.load_workbook(nome_arquivo)
ws = wb['Métricas']

cor_cabecalho = PatternFill(start_color='1F4E78', end_color='1F4E78', fill_type='solid')
fonte_cabecalho = Font(name='Arial', size=11, bold=True, color='FFFFFF')

for celula in ws[1]:
    celula.fill = cor_cabecalho
    celula.font = fonte_cabecalho

grafico = BarChart()
grafico.type = "col"
grafico.style = 10
grafico.title = "Retorno por Campanha (R$)"
grafico.y_axis.title = "Valores em R$"
grafico.x_axis.title = "Campanhas"

dados = Reference(ws, min_col=3, min_row=1, max_row=6)

categorias = Reference(ws, min_col=1, min_row=2, max_row=6)

grafico.add_data(dados, titles_from_data=True)
grafico.set_categories(categorias)

ws.add_chart(grafico, "F2")

wb.save(nome_arquivo)
print(f"🔥 Sucesso! O arquivo '{nome_arquivo}' foi gerado com métricas, design e gráfico automáticos.")
```
